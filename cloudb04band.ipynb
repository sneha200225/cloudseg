```python
import os
import rasterio

# Directory containing subdirectories with image bands (B02, B03, B04, B08)
data_directory = "/kaggle/input/cloud-cover-detection/data/train_features/"


# List of band names
band_names = [ "B04"]

# Function to check for missing data in an image
def check_missing_data(image_path):
    with rasterio.open(image_path) as src:
        missing_data = (src.read(1) == 0).sum()  # Count the number of pixels with value 0
        return missing_data

# Iterate through all subdirectories
for subdir in os.listdir(data_directory):
    subdir_path = os.path.join(data_directory, subdir)
    
    if os.path.isdir(subdir_path):
        # Initialize a dictionary to store missing data counts for each band
        missing_data_counts = {band_name: 0 for band_name in band_names}
        
        # Iterate through band names and check for missing data
        for band_name in band_names:
            # Construct the full path to the image file
            image_path = os.path.join(subdir_path, f"{band_name}.tif")
            
            # Check if the image file exists
            if os.path.exists(image_path):
                missing_data_count = check_missing_data(image_path)
                missing_data_counts[band_name] = missing_data_count
        
        # Print missing data counts for the subdirectory
        print(f"Subdirectory: {subdir}")
        for band_name, missing_count in missing_data_counts.items():
            print(f"Band: {band_name}, Missing Data Count: {missing_count}")
```

```python
import os
import rasterio

# Directory containing subdirectories with image bands (B02, B03, B08)
data_directory = "/kaggle/input/cloud-cover-detection/data/train_features/"

# List of band names
band_names = ["B04"]

# Function to get image dimensions
def get_image_dimensions(image_path):
    with rasterio.open(image_path) as src:
        height, width = src.shape
        return height, width

# Iterate through all subdirectories
for subdir in os.listdir(data_directory):
    subdir_path = os.path.join(data_directory, subdir)
    
    if os.path.isdir(subdir_path):
        # Iterate through band names and get dimensions
        for band_name in band_names:
            # Construct the full path to the image file
            image_path = os.path.join(subdir_path, f"{band_name}.tif")
            
            # Check if the image file exists
            if os.path.exists(image_path):
                height, width = get_image_dimensions(image_path)
                print(f"Subdirectory: {subdir}, Band: {band_name}, Height: {height}, Width: {width}")
            else:
                print(f"Image file not found: {image_path}")
```
